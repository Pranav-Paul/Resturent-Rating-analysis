```python
cuisines = df["Cuisines"]
```

```python
cuisines_count = cuisines.value_counts()
```

```python
Top_3_cuisines = cuisines_count.head(3)
```

```python
total = len(df)
```

```python
percentage_top_3 = (Top_3_cuisines / total) * 100
```

```python
print("top 3 cuisines:")
Top_3_cuisines
```

```python
plt.bar(percentage_top_3.index, percentage_top_3.values)
plt.xlabel("Cuisines")
plt.ylabel("Percentage")
plt.title("Top 3 Cuisines")
plt.show()
```
