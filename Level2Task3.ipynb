```python
import pandas as pd
import folium
from IPython.display import display
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Cognifyz_Task/Dataset .csv")
df.head()
```

```python
restaurants_names = df['Restaurant Name']
latitude = df["Latitude"]
longitude = df["Longitude"]
```

```python
x = df[["Latitude","Longitude"]]
num_clusters = 5
#print(df[['Latitude', 'Longitude']].isnull().sum())
```

```python
print(df.columns)
```

```python
kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)
df["cluster"] = kmeans.fit_predict(x)
```

```python
map_center = [latitude.mean(), longitude.mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)
```

```python
cluster_colors = ['red', 'blue', 'green', 'purple', 'orange']
```

```python
for index, row in df.iterrows():
  restaurant_name = row['Restaurant Name']
  latitude = row['Latitude']
  longitude = row['Longitude']
  cuisines = row['Cuisines']
  rating = row['Aggregate rating']
  cluster = row['cluster']
  popup_text = f"<b>{restaurant_name}</b><br>Cuisines: {cuisines}<br>Rating: {rating}"
  marker = folium.Marker([latitude , longitude], popup=popup_text)
  marker.add_to(restaurant_map)
```
